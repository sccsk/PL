```python
import json
```

```python
test = open('ppl.json',encoding="utf8")
Data = json.load(test)
type(Data)
```

```python
#將d裡面的內容(items)顯示出來
Data.items()
```

```python
temp = Data.get('入園人數')
temp
```

```python
#查詢月份
ask = "1月"
```

```python
for i in temp:
    if ask == i.get('月份'):
        print(ask,'入園人數',i.get('當月購票入園人數'), '人')
```

```python
allage = []
for i in temp:
    allage.append(i.get('當月購票入園人數'))
allage
```

```python
allage_enumerate = enumerate(allage, start = 1)
```

```python
allage_enumerate_sorted = sorted(allage_enumerate, key = lambda s: s[1], reverse = True)
allage_enumerate_sorted
```

```python
allage_enumerate_sorted[0][0]
```

```python
for i in allage_enumerate_sorted:
    j = i[0] - 1
    print(temp[j].get('月份'), '入園人數是', temp[j].get('當月購票入園人數'), '人')
```
